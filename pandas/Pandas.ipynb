### IsNull y Any

Is Null evalua si una columna es nulo. Devuelve una serie si se aplica a una sola columna o un dataframe si se aplica a mas de una columna.

Any evalua una columna por un True si encuentra aunque sea uno devuelve True. Devuelve un valor. 
Si se aplica a un dataframe devuelve una serie

```python
mat_test['school'].isnull() #Devuelve una serie booleana
mat_test.isnull() #Devuelve un dataframe booleano

mat_test.any() #Devuelve una serie
mat_test['age'].any() #Devuelve un valor
```

### Index

Se pueden setear y modificar los indices de un dataframe cuando se necesite. El indice normalmente se elimina de las columnas del dataframe. 

Para cambiar el indice en el dataframe en curso se debe aplicar **inplace=True**.


```python
mat_test.reset_index(inplace=True)
mat_test.set_index('school')
mat_test.set_index(['school','age'])

mat_test.set_index('school', inplace=True)
mat_test.reset_index(inplace=True)
```

### Split, Apply, Combine

* **Split:** Separo un conjunto de datos en grupos de acuerdo a un criterio especifico. Normalmente por variables categoricas.

* **Apply:** Aplico una funcion por grupos. 

* **Combine:** Uno de nuevo el dataframe

Normalmente se utiliza la operacion groupby, agregaciones y transformaciones (transform y apply)

La operacion GroupBy es liviana pues no hace procesamiento sobre todos los valores. El procesamiento se realiza al aplicar una operacion sobre el dataframe.

Pandas y NumPy manejan los nulos diferente. HAy que tener cuidado al usar funciones de uno u otro (Ejemplo: sum de Pandas vs sum de NumPy)

Una vez que se separa el dataframe se peuden aplicar funciones de agregacion (reducen el dataframe), funciones de transformacion, que mantienen el tamaño del dataframe o funciones de filtrado que reducen el tamaño por grupos.

```python
mat_test.groupby('school') #Hace el split, genera una estructura del tipo DataframeGroupBy
mat_test.groupby(['school','age'])

#Agregacion
mat_test.groupby('age').sum()
mat_test.groupby('age')['Medu'].agg(['sum','mean']) #Aplica varias funciones a una serie agrupada
mat_test.groupby('age').agg(['sum','mean']) #Aplica dos funciones a cada columna del dataframe
mat_test.groupby('age').agg({'Medu':'sum','Fedu':'mean'}) #Aplica distintas funcinoes a las columnas especificadas

#Filtering
mat_test.groupby('age').filter(lambda x: x['age'].mean() > 15)

#Transform
mat_test.groupby('age').transform() 
tbd

#Apply
tbd


```
